```python
import re
import numpy as np
import pandas as pd
import os

import seaborn as sns

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
mpl.rc('axes',labelsize=14)
mpl.rc('xtick',labelsize=10)
mpl.rc('ytick',labelsize=10)
```

```python
fhand = open('OUTCAR')
lst = list()
for line in fhand:
    line = line.rstrip().lstrip()
    if re.findall('^free energy', line):
        lst.append(line)
#print(lst)
print(*lst, sep='\n')
print(len(lst))
```

```python
numbers = []
for item in lst:
    match = re.search(r'[-+]?\d*\.\d*', item)
    #print(match.group())
    numbers.append(float(match.group()))

print(numbers)
```

```python
numbers_array=np.array(numbers)
print(numbers_array)
```

```python
DF=pd.DataFrame(numbers_array,columns=['FreeEnergyToten'])
```
